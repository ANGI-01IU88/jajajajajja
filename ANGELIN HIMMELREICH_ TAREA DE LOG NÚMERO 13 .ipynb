```python
palabra = input("Introduce una palabra: ")
vocales = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0}
for letra in palabra.lower():
    if letra in vocales:
        vocales[letra] += 1
print("Número de veces que aparece cada vocal:")
for vocal, cantidad in vocales.items():
    print(f"{vocal.upper()}: {cantidad}")
```

```python
import math
entrada = input("Introduce una lista de números separados por comas: ")
numeros = [float(num) for num in entrada.split(",")]
numeros.reverse()
media = sum(numeros) / len(numeros)
varianza = sum((x - media) ** 2 for x in numeros) / len(numeros)
desviacion_tipica = math.sqrt(varianza)
print(f"La media es: {media}")
print(f"La desviación típica es: {desviacion_tipica}")
```

```python
N = int(input("Introduce el tamaño de la lista: "))
valores = []
for i in range(N):
    valor = float(input(f"Introduce el valor #{i+1}: "))
    valores.append(valor)
valores.sort(reverse=True)
print("La lista ordenada de mayor a menor es:")
print(valores)
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        a, b = 0, 1
        for _ in range(2, n + 1):
            a, b = b, a + b
        return b
N = int(input("Introduce el tamaño de la lista: "))
valores = []
for i in range(N):
    valor = int(input(f"Introduce el valor #{i + 1} (posición en Fibonacci): "))
    valores.append(valor)
for i in range(N):
    valores[i] = fibonacci(valores[i])
print("La lista modificada con los números de Fibonacci es:")
print(valores)
```
